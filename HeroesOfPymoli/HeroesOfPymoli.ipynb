```python
import pandas as pd
```

```python
csv_file = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(csv_file)

purchase_data.head()
```

```python
total_players = purchase_data["SN"].unique()
total_players = len(total_players)

print("Total Number of Players: " + str(total_players))
```

```python
total_purchases = purchase_data["Purchase ID"].count()

num_unique_items = len(purchase_data["Item ID"].unique())

avg_purchase_price = purchase_data["Price"].sum()/total_purchases

total_revenue = purchase_data["Price"].sum()

##Create the summary dataframe
summary_df = pd.DataFrame({
    "Number of Unique Items": [num_unique_items],
    "Average Purchase Price": avg_purchase_price,
    "Total Number of Purchases": total_purchases,
    "Total Revenue": total_revenue
})

#Format the display of the $$ values
summary_df["Average Purchase Price"] = summary_df["Average Purchase Price"].astype(float).map("${:,.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].astype(float).map("${:,.2f}".format)
summary_df
```

```python
##Create a dataframe by Gender --- Calculate the percentage of each gender

gender_df = pd.DataFrame ({
    "Total Count": purchase_data["Gender"].value_counts(),
    "Percentage of Players": (purchase_data["Gender"].value_counts()/total_purchases) * 100
})
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:,.2f}%".format)
gender_df
```

```python
##Group by Gender and calculate purchase data by Gender
gender_grouped = purchase_data.groupby(['Gender'])

purchase_count = gender_grouped["Purchase ID"].count()
total_purchase_value = gender_grouped["Price"].sum()
avg_purchase_price = total_purchase_value/purchase_count

avg_per_person = total_purchase_value/gender_grouped["SN"].nunique()

gender_group_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase_price,
    "Total Purchase Value": total_purchase_value,
    "Avg Total Purchase per Person": avg_per_person
})
gender_group_df["Average Purchase Price"] = gender_group_df["Average Purchase Price"].map("${:,.2f}".format)
gender_group_df["Total Purchase Value"] = gender_group_df["Total Purchase Value"].map("${:,.2f}".format)
gender_group_df["Avg Total Purchase per Person"] = gender_group_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
gender_group_df
```

```python
# Create bins in which to place values based upon players ages
bins = [0, 9, 14, 19, 24, 29, 34, 39, 110]

# Create labels for these bins
group_labels = ["Under 10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

##Add and age group column to the original data frame
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels = group_labels)
purchase_data.head()
```

```python
##Create a dataframe by Age --- Calculate the percentage of each Age group

age_group = purchase_data.groupby("Age Group") 

age_group_df = pd.DataFrame({
    "Total Count": age_group["Purchase ID"].count(),
    "Percentage of Players": (age_group["Purchase ID"].count()/total_purchases) * 100
})
age_group_df["Percentage of Players"] = age_group_df["Percentage of Players"].map("{:,.2f}%".format)

age_group_df

```

```python
##Puchasing Analysis by Age groups

purchase_count = age_group["Purchase ID"].count()
total_purchase_value = age_group["Price"].sum()
avg_purchase_price = total_purchase_value/purchase_count

avg_per_person = total_purchase_value/age_group["SN"].nunique()

age_grouped_df = pd.DataFrame({
    "Purchase Count": purchase_count,
    "Average Purchase Price": avg_purchase_price,
    "Total Purchase Value": total_purchase_value,
    "Avg Total Purchase per Person": avg_per_person
})
age_grouped_df["Average Purchase Price"] = age_grouped_df["Average Purchase Price"].map("${:,.2f}".format)
age_grouped_df["Total Purchase Value"] = age_grouped_df["Total Purchase Value"].map("${:,.2f}".format)
age_grouped_df["Avg Total Purchase per Person"] = age_grouped_df["Avg Total Purchase per Person"].map("${:,.2f}".format)
age_grouped_df
```
